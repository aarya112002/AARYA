```python
def pizza_menu():
    print("\nüçï Welcome to Aarya Pizza House üçï")
    print("1. Margherita Pizza - ‚Çπ200")
    print("2. Farmhouse Pizza  - ‚Çπ300")
    print("3. Paneer Pizza     - ‚Çπ250")
    print("4. Veggie Pizza     - ‚Çπ220")
    print("5. Exit")

total_bill = 0

while True:
    pizza_menu()
    choice = int(input("Enter your choice (1-5): "))

    if choice == 1:
        qty = int(input("Enter quantity: "))
        total_bill += 200 * qty
        print("Margherita Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 2:
        qty = int(input("Enter quantity: "))
        total_bill += 300 * qty
        print("Farmhouse Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 3:
        qty = int(input("Enter quantity: "))
        total_bill += 250 * qty
        print("Paneer Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 4:
        qty = int(input("Enter quantity: "))
        total_bill += 220 * qty
        print("Veggie Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 5:
        print("\nThank you for ordering üôè")
        print("Your Final Total Bill: ‚Çπ", total_bill)
        break

    else:
        print("Invalid choice, Please choose a number between 1 and 5.")
```
