```python
def pizza_menu():
    print("\nüçï Welcome to Aarya Pizza House üçï")
    print("1. Margherita Pizza - ‚Çπ200")
    print("2. Farmhouse Pizza  - ‚Çπ300")
    print("3. Paneer Pizza     - ‚Çπ250")
    print("4. Veggie Pizza     - ‚Çπ220")
    print("5. Checkout & Pay")
    print("6. Exit without ordering")
    print("7. Remove Last Order")

total_bill = 0
last_order_price = 0
last_item_name = ""
last_item_qty = 0

while True:
    pizza_menu()

    try:
        choice = int(input("Enter your choice (1-7): "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if choice == 1:
        last_item_qty = int(input("Enter quantity: "))
        last_order_price = 200 * last_item_qty
        last_item_name = "Margherita Pizza"
        total_bill += last_order_price
        print("Margherita Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 2:
        last_item_qty = int(input("Enter quantity: "))
        last_order_price = 300 * last_item_qty
        last_item_name = "Farmhouse Pizza"
        total_bill += last_order_price
        print("Farmhouse Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 3:
        last_item_qty = int(input("Enter quantity: "))
        last_order_price = 250 * last_item_qty
        last_item_name = "Paneer Pizza"
        total_bill += last_order_price
        print("Paneer Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 4:
        last_item_qty = int(input("Enter quantity: "))
        last_order_price = 220 * last_item_qty
        last_item_name = "Veggie Pizza"
        total_bill += last_order_price
        print("Veggie Pizza added")
        print("Current Total Bill: ‚Çπ", total_bill)

    elif choice == 7:
        if total_bill == 0:
            print("No order to remove.")
        else:
            total_bill -= last_order_price
            print(f"{last_item_name} removed ‚ùå")
            print("Current Total Bill: ‚Çπ", total_bill)
            last_order_price = 0
            last_item_name = ""
            last_item_qty = 0

    elif choice == 5:
        if total_bill == 0:
            print("You have not ordered anything.")
            continue

        print("\nüí∞ Total Amount to Pay: ‚Çπ", total_bill)
        print("Select Payment Mode:")
        print("1. Cash")
        print("2. UPI")
        print("3. Card")

        payment = int(input("Enter payment choice (1-3): "))

        if payment == 1:
            payment_mode = "Cash"
        elif payment == 2:
            payment_mode = "UPI"
        elif payment == 3:
            payment_mode = "Card"
        else:
            payment_mode = "Unknown"

        # üßæ INVOICE
        print("\nüßæ --------- INVOICE ---------")
        print("Pizza Name :", last_item_name)
        print("Quantity   :", last_item_qty)
        print("Price      : ‚Çπ", last_order_price)
        print("Total Bill : ‚Çπ", total_bill)
        print("Payment    :", payment_mode)
        print("-----------------------------")

        print("\n Order Confirmed!")
        print("Thank you for ordering üôè Visit Again!")
        break

    elif choice == 6:
        print("\nThank you! Program exited üëã")
        break

    else:
        print("Invalid choice, Please choose between 1 and 7.")
```
